```python
from sqlalchemy import create_engine
import pandas as pd
import os  

# Crear conector
conn = create_engine(os.path.join('sqlite:///','data', 'nba.db'))
```

```python
conn.table_names()
```

```python
# funcion de consultas
def consulta(query,conn):
    return pd.read_sql_query(query, con=conn)
```

```python
# ejemplo
query = """
SELECT * FROM player
"""
consulta(query,conn)
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos mediante código **SQL**. Para cumplir con este objetivo debe resolver las siguientes problemáticas:

1. Mostrar las primeras 5 filas

```python
query = """
SELECT * FROM player 
LIMIT 5;
"""
consulta(query,conn)
```

2. Seleccionar los valores únicos de la columna `position`.

```python
query = """
SELECT DISTINCT position FROM player 

"""
consulta(query,conn)
```

3. Seleccionar y cambiar el nombre de la columna `name` por `nombre` 

```python
query = """
SELECT 
name as nombre 
FROM player 

"""
consulta(query,conn)
```

4. Determinar el tiempo (en años) de cada jugador en su posición

```python
#Basta con sacar la diferencia entre los años
query = """
SELECT 
name, position, year_start, year_end,  year_end - year_start AS años
FROM player
"""
consulta(query,conn)
```

5. Encontrar el valor máximo de la columna `weight` por cada  valor de la columna `position`

```python
query = """
SELECT 
position, MAX(weight)  
FROM player GROUP BY position

"""
consulta(query,conn)
```

6. Encontrar el total de jugadores por cada valor de la columna `year_start`

```python
query = """
SELECT 
year_start, COUNT(name)  
FROM player GROUP BY year_start

"""
consulta(query,conn)
```

7. Encontrar el valor mínimo, máximo y promedio de la columna `weight` por cada valor de la columnas `college`

```python
query = """
SELECT 
college, MIN(weight),MAX(weight),AVG(weight)  
FROM player GROUP BY college
"""

consulta(query,conn)
```

8. Filtrar por aquellos jugadores que cumplan con :
* Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
* Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
* Para la columna `college` tienen un valor distinto de `Duke University`
**Entendí que era una tabla por cada uno**


```python
#Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
query = """
SELECT 
name, year_start
FROM player 
WHERE year_start BETWEEN 1990 AND 2000
"""

consulta(query,conn)
```

```python
#Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
query = """
SELECT 
name, position
FROM player 
WHERE position = 'C' OR position = 'C-F' OR position = 'F-C'
"""

consulta(query,conn)
```

```python
#Para la columna `college` tienen un valor distinto de `Duke University`
query = """
SELECT 
name, college
FROM player 
WHERE NOT college = 'Duke University'
"""

consulta(query,conn)
```

9. Crear dos conjuntos de datos y juntarlos en una misma *query*. Las condiciones de cada uno de los cojunto de datos son:

* **df1**:
    * Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
    * Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
   
* **df2**:
    * Para la columna `year_end` tienen un valor menor a 2000
    * Para la columna `position` tienen un valor de `G`o `F` 

```python
query1 = """
SELECT 
name, year_start, year_end, position
FROM player 
WHERE year_start BETWEEN 1990 AND 2000 AND (position = 'C' OR position = 'C-F' OR position = 'F-C')
"""

consulta(query1,conn)
```

```python
query2 = """
SELECT 
name,year_start, year_end, position
FROM player 
WHERE year_end <  2000 AND (position = 'G' OR position = 'F' )
"""

consulta(query2,conn)
```

```python
df1 = pd.read_sql_query(query1, conn)#se crea el dataframe de df1
df2 = pd.read_sql_query(query2, conn)#se crea el dataframe de df2
```
